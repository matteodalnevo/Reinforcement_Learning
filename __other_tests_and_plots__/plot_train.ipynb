```python
import numpy as np
from  tqdm import trange
import matplotlib.pyplot as plt 
```

# Base Test

```python
# Load the data
data = np.load("./data/train/data_baseline.npz")

max_steps_baseline = data['max_steps']
n_fruit_eaten_baseline = np.ones(100) * data['n_fruit_eaten_ep']
n_wall_hit_baseline = np.ones(100) * data['n_wall_hit_ep']
n_body_hit_baseline = np.ones(100) * data['n_body_hit_ep']
rewards_baseline = np.ones(100) * data['rewards_ep']

data = np.load("./data/train/data_a2c.npz")

max_steps_a2c = data['max_steps']
n_fruit_eaten_a2c = data['n_fruit_eaten_ep']
n_wall_hit_a2c = data['n_wall_hit_ep']
n_body_hit_a2c = data['n_body_hit_ep']
rewards_a2c = data['rewards_ep']

data = np.load("./data/train/data_ddqn.npz")

max_steps_ddqn = data['max_steps']
n_fruit_eaten_ddqn = data['n_fruit_eaten_ep']
n_wall_hit_ddqn = data['n_wall_hit_ep']
n_body_hit_ddqn = data['n_body_hit_ep']
rewards_ddqn = data['rewards_ep']

data = np.load("./data/train/data_ppo.npz")

max_steps_ppo = data['max_steps']
n_fruit_eaten_ppo = data['n_fruit_eaten_ep']
n_wall_hit_ppo = data['n_wall_hit_ep']
n_body_hit_ppo = data['n_body_hit_ep']
rewards_ppo = data['rewards_ep']

# Plotting
plt.figure(figsize=(12, 8))

# Plotting n_fruit_eaten
plt.subplot(2, 2, 1)
plt.plot(n_fruit_eaten_a2c, label='A2C')
plt.plot(n_fruit_eaten_ddqn, label='DDQN')
plt.plot(n_fruit_eaten_ppo, label='PPO')
plt.plot(n_fruit_eaten_baseline, linestyle='--', label='BL', color='red')
plt.title('Average Fruits Eaten per Epoch')
plt.xlabel('Epochs')
plt.ylabel('Average Fruits Eaten')
plt.legend()
plt.grid(True)

# Plotting n_wall_hit
plt.subplot(2, 2, 2)
plt.plot(n_wall_hit_a2c, label='A2C')
plt.plot(n_wall_hit_ddqn, label='DDQN')
plt.plot(n_wall_hit_ppo, label='PPO')
plt.plot(n_wall_hit_baseline, linestyle='--', label='BL', color='red')
plt.title('Average Wall Hits per Epoch')
plt.xlabel('Epochs')
plt.ylabel('Average Wall Hits')
plt.legend()
plt.grid(True)

# Plotting n_body_hit
plt.subplot(2, 2, 3)
plt.plot(n_body_hit_a2c, label='A2C')
plt.plot(n_body_hit_ddqn, label='DDQN')
plt.plot(n_body_hit_ppo, label='PPO')
plt.plot(n_body_hit_baseline, linestyle='--', label='BL', color='red')
plt.title('Average Body Hits per Epoch')
plt.xlabel('Epochs')
plt.ylabel('Average Body Hits')
plt.legend()
plt.grid(True)

# Plotting rewards
plt.subplot(2, 2, 4)
plt.plot(rewards_a2c, label='A2C')
plt.plot(rewards_ddqn, label='DDQN')
plt.plot(rewards_ppo, label='PPO')
plt.plot(rewards_baseline, linestyle='--', label='BL', color='red')
plt.title('Average Rewards per Epoch')
plt.xlabel('Epochs')
plt.ylabel('Average Rewards')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
